# IOGA and Differential Equations

Chemical engineers use IOGA to develop differential equations that allow them to predict how much of something is in their system. Differential equations are equations that relate a one or more functions and their derivatives. A derivative is the rate some variable changes with respect to another.

IOGA is an acronym for how chemical engineers track the contents of their system. IOGA is also known as material balances, mass balances, energy balances, or population balances. Ordinary differential equations (ODEs) are the result when the system is not at steady state, so that the accumulation term may be non-zero. Other fields may call development of these ODEs as compartment analysis or population dynamics.

$I - O + G = A$

$I$ = How much comes into the system

$O$ = How much is going out from the system

$G$ = How much is generated/consumed in the system

$A$ = How much is accumulated in the system

````python
import numpy as np                      #Numerical python (np.zeros, np.linspace)
from scipy.integrate import solve_ivp   #the function used to solve the differential equation with an initial condition
import math                             #additional functions (nothing used in this template)
import matplotlib.pyplot as plt         #MATLAB like plotting package (plt.figure, plt.plot)
````

## Write a function that contains the ODE

This function named <font color=blue>ODEfun</font>  returns the value of the derivative of dydt so that it can be integrated using the python function solve_ivp.
Solve_ivp requires that you input a function that has inputs of the independent variable (t) and the dependent variables $C_A$, $C_B$, $C_C$ are passed out of the function as the
array y.  ODEfun(independent variable, array of dependent variables) (e.g., for dy/dt = ODEfun(t,y))


````python
#This is the ODEfun(independent variable, array of dependent variables) (e.g., for a system of ODE dy/dt ODEfun(t,y))
def  ODEfun(t,y):      #fun(t,y) and return dy/dt
  # Set parameters also known as rate constants. Use comments to indicate
  # the units for each after giving the values
  a1 = 0.6 # M/hr
  b1 = 0.6 # 1/Mhr
  c1 = 0.1 # M/hr
  a2 = 0.2 # M/hr
  c2 = 0.2 # M/hr
  c3 = 0.2 # M/hr

  # Assign y's to the correct species
  # If there are more or fewer variables, then update the structure as
  # variablename = y[i] for i = 0 to N-1; Python starts counting at 0, not 1
  C_A = y[0] # This unpacks the first element of the array named y
  C_B = y[1] # This unpacks the next element of the array named y
  C_C = y[2] # This unpacks the next element of the array named y

  # If you need any algebraic relationships, type them in here
  T = C_A*C_B # unnecessary example

  # Define the ODEs
  # Note: change this list to match the differential equations. Make sure
  # that your variable names match those listed above. Do not include
  # spaces in any of the variable names and do not use / for dC/dt.
  # Note: the indentation has to be consistent or Python will have an error
  dC_Adt = a1-b1*T-c1
  dC_Bdt = a2-b1*C_A*C_B-c2
  dC_Cdt = b1*C_A*C_B-c3

  dydt = [dC_Adt, dC_Bdt, dC_Cdt]      #This makes an array of the values of the derivatives.
  return dydt        #This sends the values of the derivatives out to the Jupyter Notebook
````

## Set up the tspan, initial conditions, and call the ODE solver

````python
# Set the time range
# Note: these times may change for each problem
tinitial = 0 # h
tfinal = 6 # h
numberOfTimes = 101
tspan = (tinitial,tfinal)
teval = np.linspace(tinitial, tfinal, numberOfTimes) # Range for the independent variable t

# Set the initial conditions for the dependent variables
# Note: the Ci initial condition vector has three entries because we have three ODEs.
# If only one ODE, then Ci = [firstvalue];
# If two ODEs, the Ci = [ firstvalue, secondvalue];
Ci=[6, 3, 0] # C_A at time 0=6 M, C_B at time 0=3 M, C_C at time 0 = 0 M

# Use the ODE solver to solve the equations
#The inputs to solve_ivp are the equations in the function ODEfun, the integration start&stop,
#and the initial values of Ci
sol = solve_ivp(ODEfun,tspan,Ci,t_eval=teval)
print (sol)
# sol is the vector that stores the calculated values for each timestep
# ODEfun calls the function that defines the system of equations, which we need to update if we want to solve a different system of equations


# We pull the concentrations out from the sol vector to explore the solution and graph it
# Note: the variable names on the left hand sides of the equations can be renamed.
# Note: the 0, 1, 2 are for the three variables. If you have N variables, you only have N outputs for sol.y[i] for i = 0 to N-1.
t = sol.t
C_A=sol.y[0]
C_B=sol.y[1]
C_C=sol.y[2]
````

````python
#plot the data
plt.figure(figsize=(10,6))

plt.plot(sol.t,C_A,'b-',label='C_A')
plt.plot(sol.t,C_B,'k:',label='C_B')
plt.plot(sol.t,C_C,'g--',label='C_C')

plt.xlabel('Time (h)')
plt.ylabel('Concentration (M)')

plt.legend()
plt.title('Concentration vs. t')
````
